```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"
```

```python
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
def getGeneration(ndex):
    ndex_int = int(ndex[1:])
    if ndex_int >= 1 and ndex_int <= 151:
        return '1'
    elif ndex_int >= 152 and ndex_int <= 251:
        return '2'
    elif ndex_int >= 252 and ndex_int <= 386:
        return '3'
    elif ndex_int >= 387 and ndex_int <= 493:
        return '4'
    elif ndex_int >= 494 and ndex_int <= 649:
        return '5'
    elif ndex_int >= 650 and ndex_int <= 721:
        return '6'
    elif ndex_int >= 722 and ndex_int <= 809:
        return '7'
    elif ndex_int >= 810 and ndex_int <= 905:
        return '8'
```

```python
types = ['Grass', 'Fire', 'Water', 'Poison', 'Fairy', 'Steel', 'Rock', 'Fighting', 'Psychic', 'Dark', 
         'Ghost', 'Ice', 'Flying', 'Bug', 'Ground', 'Electric', 'Normal', 'Dragon']
poke_content=soup.find(id='mw-content-text')
poke_tables=poke_content.find_all('td')
i = 1
while i < len(poke_tables) - 33:
    # regional dex instead of kdex to generalize regional dex
    rdex = poke_tables[i].text.strip()
    ndex = poke_tables[i + 1].text.strip()
    name = poke_tables[i + 2].text.strip()
    gen = getGeneration(ndex)
    scrape_time = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    type1 = poke_tables[i + 3].text.strip()
    if poke_tables[i + 4].text.strip() in types:
        type2 = poke_tables[i + 4].text.strip()
        print(f'Pokemon {ndex} {name} is a {type1} & {type2} Pokemon from Generation {gen} scraped {scrape_time}')
        i += 5
    else:
        print(f'Pokemon {ndex} {name} is a {type1} Pokemon from Generation {gen} scraped {scrape_time}')
        i += 4
```

```python
poke_json = []
i = 1
while i < len(poke_tables) - 33:
    # regional dex instead of kdex to generalize regional dex
    rdex = poke_tables[i].text.strip()
    ndex = poke_tables[i + 1].text.strip()
    name = poke_tables[i + 2].text.strip()
    gen = getGeneration(ndex)
    scrape_time = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    type1 = poke_tables[i + 3].text.strip()
    if poke_tables[i + 4].text.strip() in types:
        type2 = poke_tables[i + 4].text.strip()
        poke_json.append({
            "rdex": rdex,
            "ndex": ndex,
            "name": name,
            "type1": type1,
            "type2": type2,
            "generation": gen,
            "scrape_time": scrape_time
        })
        i += 5
    else:
        poke_json.append({
            "rdex": rdex,
            "ndex": ndex,
            "name": name,
            "type1": type1,
            "generation": gen,
            "scrape_time": scrape_time
        })
        i += 4
        
poke_json
```

```python
with open('esquivel_chua_pokemon.json', 'w') as fp:
    json.dump(poke_json, fp, indent = 4)
```
